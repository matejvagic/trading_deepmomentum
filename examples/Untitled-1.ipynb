```python
import pandas as pd
import numpy as np
from numpy import array
```

```python
df = pd.read_csv('Stock-market-forecasting-master-1/data/SPY_returns_new.csv',parse_dates=True)
df.head()
```

```python
df = df.loc[:, (df != df.iloc[0]).any()] 
df.shape
```

```python
from tensorflow import keras 
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
from sklearn.preprocessing import StandardScaler
```

```python
df_only_ret_standardised = df

scalerReturn = StandardScaler()
scalerCR = StandardScaler()
#scalerOR = StandardScaler()

scalerReturn.fit(df_only_ret_standardised["returns"].to_numpy().reshape(-1, 1))
scalerCR.fit(df_only_ret_standardised["cr"].to_numpy().reshape(-1, 1))
#scalerOR.fit(df_only_ret_standardised["or"].to_numpy().reshape(-1, 1))


df_only_ret_standardised["stand_returns"] = scalerReturn.transform(df_only_ret_standardised["returns"].to_numpy().reshape(-1, 1))
df_only_ret_standardised["stand_cr"] = scalerCR.transform(df_only_ret_standardised["cr"].to_numpy().reshape(-1, 1))
#df_only_ret_standardised["stand_or"] = scalerOR.transform(df_only_ret_standardised["or"].to_numpy().reshape(-1, 1))


#df_only_ret_standardised = df_only_ret_standardised[["stand_ir", "stand_cr", "stand_or"]]
df_only_ret_standardised  = df_only_ret_standardised.drop(["cr", "returns"], axis=1)

df_only_ret_standardised.head()
```

```python
cols = df_only_ret_standardised.columns.tolist()
cols = cols[-2:] + cols[:-2]
cols
```

```python
df_only_ret_standardised = df_only_ret_standardised[cols]
df.head()
```

```python
look_back = 200
def load_data(stock, look_back):
    data_raw = stock.values
    data = []
    
    # create all possible sequences of length look_back
    for index in range(len(data_raw) - look_back): 
        data.append(data_raw[index: index + look_back])
    
    data = np.array(data);
    test_set_size = int(np.round(0.2*data.shape[0]));
    train_set_size = data.shape[0] - (test_set_size);
    
    x_train = data[:train_set_size,:,:]
    y_train = data[:train_set_size,0,:1]
    
    x_test = data[train_set_size:,:,:]
    y_test = data[train_set_size:,0,:1]
    
    return [x_train, y_train, x_test, y_test]

x_train, y_train, x_test, y_test = load_data(df_only_ret_standardised, look_back)
print(x_train.shape)
print(y_train.shape)
```

```python
df_only_ret_standardised.tail()
```

```python
y_train_class = scalerReturn.inverse_transform(y_train)
```

```python
y_train_class[y_train_class >=0] = 1
y_train_class[y_train_class <0] = 0
```

```python
y_train_class
```

```python
from keras.layers import Dropout
from keras.layers import ConvLSTM2D

n_steps = look_back


n_features = 8
x_train = x_train.reshape((x_train.shape[0], x_train.shape[1], n_features))

# define model
model = Sequential()
model.add(ConvLSTM2D(1, (5,5), input_shape=(n_steps, n_features)))
model.add(Dropout(0.1))
model.add(LSTM(50,   input_shape=(n_steps, n_features),))


model.add(Dropout(0.1))
model.add(Dense(1))
model.compile(loss='bce',
                          metrics=['accuracy'])
```

```python
model.fit(x_train, y_train_class, epochs=100, validation_split=0.2, batch_size=512)
```

```python
preds_class= np.transpose(model.predict(x_train, verbose=0))
```

```python
preds_class
```

```python
out = 0
positive = 0
negative = 0
for i in range(len(preds_class[0])):
  if (preds_class[0][i]>=0.5):
    positive+=1
  else:
    negative+=1
  if preds_class[0][i] >= 0.5 and  np.transpose(y_train_class)[0][i] >=0.5 or preds_class[0][i] < 0.5 and np.transpose(y_train_class)[0][i] < 0.5:
    out += 1
  
print(f'Positive: {positive}')
print(f'Negative: {negative}')
print(out/len(preds_class[0]))
```
